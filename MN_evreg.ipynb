```python
# Dependencies and Setup
import pandas as pd
from pathlib import Path

# File to Load (Remember to Change These)
evreg_data_to_load = Path("Resources/MN_EV_Registrations.csv")
zipcode_data_to_load = Path("Resources/uszips.csv")

# Read School and Student Data File and store into Pandas DataFrames
evreg_data = pd.read_csv(evreg_data_to_load)
zipcode_data = pd.read_csv(zipcode_data_to_load)

zipcode_data.dtypes
# Combine the data into a single dataset.
#evreg_data_complete = pd.merge(evreg_data, zipcode_data, how="left", on="ZIP Code", sort=True)
#evreg_data_complete.head()
```

```python
# Calculate the t
mn_zipcode_df = zipcode_data[(zipcode_data["state_name"] == "Minnesota")]
mn_zipcode_df = mn_zipcode_df[['zip','lat','lng','city']]
mn_zipcode_df.rename(columns={'zip':'ZIP Code'}, inplace=True)
mn_zipcode_df['ZIP Code'] = mn_zipcode_df['ZIP Code'].astype('str')
mn_zipcode_df
```

```python
from datetime import datetime as dt
#campaign_df["launch_date"] = pd.to_datetime(campaign_df["launch_date"].dt.date,format='%Y-%m-%d')
evreg_data["Registration Date"] = pd.to_datetime(evreg_data["Registration Date"],format='%m/%d/%Y')
evreg_data.info()
```

```python
evreg_2022_df = evreg_data[(evreg_data["Registration Date"].dt.year == 2022)]
evreg_2022_counts = evreg_2022_df['ZIP Code'].value_counts()
evreg_2022_counts_df = pd.DataFrame(evreg_2022_counts)

evreg_2022_counts_df
```

```python
evreg_2023_df = evreg_data[(evreg_data["Registration Date"].dt.year == 2023)]
evreg_2023_counts = evreg_2023_df['ZIP Code'].value_counts()
evreg_2023_counts_df = pd.DataFrame(evreg_2023_counts)
evreg_2023_counts_df
```

```python
evreg_2021_df = evreg_data[(evreg_data["Registration Date"].dt.year == 2021)]
evreg_2021_counts = evreg_2021_df['ZIP Code'].value_counts()
evreg_2021_counts_df = pd.DataFrame(evreg_2021_counts)
evreg_2021_counts_df
```

```python
evreg_2022_counts_zipmerge = evreg_2022_counts_df.merge(mn_zipcode_df, on='ZIP Code')
evreg_2022_counts_zipmerge
```

```python
evreg_2023_counts_zipmerge = evreg_2023_counts_df.merge(mn_zipcode_df, on='ZIP Code')
evreg_2023_counts_zipmerge
```

```python
evreg_2021_counts_zipmerge = evreg_2021_counts_df.merge(mn_zipcode_df, on='ZIP Code')
evreg_2021_counts_zipmerge
```

```python
evreg_2022_counts_zipmerge.info()
evreg_2023_counts_zipmerge.info()
evreg_2021_counts_zipmerge.info()
```

```python
import json
from json import loads, dumps
from pprint import pprint

geojson = {"type": "FeatureCollection", "features": []}

for _, row in evreg_2021_counts_zipmerge.iterrows():
    feature = {"type": "Feature", "geometry": {"type": "Point", "coordinates": [row['lng'], row['lat']]}, "properties": {"city": row['city'], "reg_count": row['count']}}
    geojson['features'].append(feature)

with open('evreg_2021.geojson', 'w') as fp:
    json.dump(geojson, fp, indent=4)  

```

```python
geojson = {"type": "FeatureCollection", "features": []}

for _, row in evreg_2023_counts_zipmerge.iterrows():
    feature = {"type": "Feature", "geometry": {"type": "Point", "coordinates": [row['lng'], row['lat']]}, "properties": {"city": row['city'], "reg_count": row['count']}}
    geojson['features'].append(feature)

with open('evreg_2023.geojson', 'w') as fp:
    json.dump(geojson, fp, indent=4)  
```
