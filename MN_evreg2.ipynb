```python
# Dependencies and Setup
import pandas as pd
from pathlib import Path

# File to Load 
evreg2023_data_to_load = Path("Resources/2023_EV_Registration_Data.xlsx")
evreg2022_data_to_load = Path("Resources/2022_EV_Registration_Data.xlsx")
evreg2021_data_to_load = Path("Resources/2021_EV_Registration_Data.xlsx")

# Read files and store into Pandas DataFrames
evreg2023_data = pd.read_excel(evreg2023_data_to_load)
evreg2022_data = pd.read_excel(evreg2022_data_to_load)
evreg2021_data = pd.read_excel(evreg2021_data_to_load)

# zipcode_data.dtypes
# Combine the data into a single dataset.
#evreg_data_complete = pd.merge(evreg_data, zipcode_data, how="left", on="ZIP Code", sort=True)
#evreg_data_complete.head()
evreg2021_data.info()
evreg2022_data.info()
evreg2023_data.info()
```

```python
# Calculate the
evreg2021_data['County_FIPS'] = evreg2021_data['County_FIPS'].astype(str).str.replace('.0', '', regex=False)         
evreg2021_data_df = evreg2021_data.dropna()
evreg2021_data_df.head()
```

```python
# Calculate the
evreg2022_data['County_FIPS'] = evreg2022_data['County_FIPS'].astype(str).str.replace('.0', '', regex=False)         
evreg2022_data_df = evreg2022_data.dropna()
evreg2022_data_df.head()
```

```python
# Calculate the
evreg2023_data['County_FIPS'] = evreg2023_data['County_FIPS'].astype(str).str.replace('.0', '', regex=False)         
evreg2023_data_df = evreg2023_data.dropna()
evreg2023_data_df.head()
```

```python
evreg2021_counts = evreg2021_data_df['County_FIPS'].value_counts()

evreg2021_counts_df = pd.DataFrame(evreg2021_counts)
evreg2021_counts_df.reset_index(inplace=True)
evreg2021_counts_df
```

```python
evreg2022_counts = evreg2022_data_df['County_FIPS'].value_counts()

evreg2022_counts_df = pd.DataFrame(evreg2022_counts)
evreg2022_counts_df.reset_index(inplace=True)
evreg2022_counts_df
```

```python
evreg2023_counts = evreg2023_data_df['County_FIPS'].value_counts()
evreg2023_counts_df = pd.DataFrame(evreg2023_counts)
evreg2023_counts_df.reset_index(inplace=True)
evreg2023_counts_df
```

```python
evreg2021_counts_df.to_csv('MN_2021_ev_reg_by_county.csv', index=False)
evreg2022_counts_df.to_csv('MN_2022_ev_reg_by_county.csv', index=False)
evreg2023_counts_df.to_csv('MN_2023_ev_reg_by_county.csv', index=False)
```
