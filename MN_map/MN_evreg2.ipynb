```python
# Dependencies and Setup
import pandas as pd
from pathlib import Path

# File to Load 
evreg2023_data_to_load = Path("Resources/2023_EV_Registration_Data.xlsx")

# Read files and store into Pandas DataFrames
evreg2023_data = pd.read_excel(evreg2023_data_to_load)
evreg2023_data.info()
```

```python
# Convert county code to string and drop trailing decimal
evreg2023_data['County_FIPS'] = evreg2023_data['County_FIPS'].astype(str).str.replace('.0', '', regex=False)      
# drop any null county codes   
evreg2023_data_df = evreg2023_data.dropna()
evreg2023_data_df.head()
```

```python
# get vehicle registration counts by county code in dataframe
evreg2023_counts = evreg2023_data_df['County_FIPS'].value_counts()
evreg2023_counts_df = pd.DataFrame(evreg2023_counts)
evreg2023_counts_df.reset_index(inplace=True)
evreg2023_counts_df.tail()
```

```python
# write counts data to output file
evreg2023_counts_df.to_csv('Resources/MN_2023_ev_reg_by_county.csv', index=False)
```
